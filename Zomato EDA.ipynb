```python
df_country = pd.read_excel(r"C:\Users\tijo.thomas\OneDrive - edelweisstokio.in\Desktop\Python\Exploratory Data Analysis\Zomatodataset\Country-Code.xlsx")
df_country
```

```python
final_df = pd.merge(df, df_country , on = 'Country Code', how ='left')
final_df
```

```python
#To check data types
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names  = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
##plotting Pie Chart
plt.pie(country_val[:3], labels = country_names[:3] , autopct= '%1.2F%%')
```

```python
#Observation : Zomato maximum records are from India , Usa and UK respectively
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Ratings count'})
```

```python
ratings
```

```python
##Observations
1 . When rating is from 4.5 to 4.9 ---> Excellent
2 . When rating is from 4. to 3.4 ---> Very Good
3 . When rating is from 3.5 to 3.9 ---> Good


```

```python
ratings.head()
```

```python
plt.rcParams["figure.figsize"] = (20,3)
sns.barplot(x='Aggregate rating' , y='Ratings count', data = ratings)
```

```python
sns.barplot(x='Aggregate rating' , y='Ratings count', data = ratings , hue = 'Rating color' , palette = ['white','red','orange','yellow','green' ,'blue'])
```

```python
#Observation
1. Not rated count is very high
2. Maximum number of ratings are from 2.5 to 3.4
```

```python
##Count Plot
#Frequency of colors
sns.countplot(x = 'Rating color', data = ratings ,  palette = ['white','red','orange','yellow','green' ,'blue'])
```
